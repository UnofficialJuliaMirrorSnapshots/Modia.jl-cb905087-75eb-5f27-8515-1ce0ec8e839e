```julia
using Modia
using Modia.Electric
using ModiaMath.plot
```

```julia
@model LPfilter begin
  R=Resistor(R=100.0)
  C=Capacitor(C=2.5E-3)
  V=ConstantVoltage(V=10.0)
@equations begin
  connect(V.p, R.p)
  connect(R.n, C.p)
  connect(C.n, V.n)
  end
end;
```

```julia
result = simulate(LPfilter, 2);
```

```julia
plot(result, ("C.v", "V.v"))
```
